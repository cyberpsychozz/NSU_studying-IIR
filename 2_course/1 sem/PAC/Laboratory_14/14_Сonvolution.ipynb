## Свёртка
Свертка (англ. convolution) — это операция, показывающая «схожесть» одной функции с отражённой и сдвинутой копией другой. Понятие свёртки обобщается для функций, определённых на группах, а также мер. 

<img src="/home/cyberpsychoz/NSU_studying-IIR/2_course/PAC/images/LessonsII/convolution.jpeg" alt="Conv" height=40% width=40%>

$$O_{i, j} = f(K*I_{i-s:i+s, j-s:j+s})$$

$K$ - ядро свёртки  
$I$ - входной тезнор  
$I_{i-s:i+s, j-s:j+s}$ - срез тензора размера ядра свёртки с центром в точке (i, j)  
$O_{i, j}$ - результат свёртки (значение выходного тензора в точке (i, j))  
$f$ - функция активации

Слой свёртки (convolutional layer) — это основной блок свёрточной нейронной сети. Слой свёртки включает в себя для каждого канала свой фильтр, ядро свёртки которого обрабатывает предыдущий слой по фрагментам (суммируя результаты поэлементного произведения для каждого фрагмента). Весовые коэффициенты ядра свёртки (небольшой матрицы) неизвестны и устанавливаются в процессе обучения.

function im2col(image, kernel_height, kernel_width, stride=1, padding=0, dilation=1):
    # image: 2D array [height, width]
    # Добавляем padding
    padded_image = pad(image, padding)  # Функция pad добавляет нули по краям

    # Вычисляем размеры выхода
    out_height = floor((height + 2*padding - dilation*(kernel_height - 1) - 1) / stride + 1)
    out_width = floor((width + 2*padding - dilation*(kernel_width - 1) - 1) / stride + 1)
    
    # Количество столбцов = out_height * out_width
    num_columns = out_height * out_width
    
    # Размер столбца = kernel_height * kernel_width
    column_size = kernel_height * kernel_width
    
    # Создаём матрицу [column_size, num_columns]
    columns = zeros(column_size, num_columns)
    
    # Индекс для столбца
    col_idx = 0
    
    # Цикл по позициям ядра
    for i in range(0, out_height):
        for j in range(0, out_width):
            # Начало патча
            row_start = i * stride
            col_start = j * stride
            
            # Извлекаем патч и развёртываем в столбец
            patch = padded_image[row_start : row_start + kernel_height*dilation : dilation,
                                 col_start : col_start + kernel_width*dilation : dilation]
            columns[:, col_idx] = flatten(patch)  # Разворачиваем в вектор
            col_idx += 1
    
    return columns
