# Архитектуры нейронных сетей

## AlexNet
Архитектура AlexNet состоит из пяти свёрточных слоёв, между которыми располагаются pooling-слои и слои нормализации, а завершают нейросеть три полносвязных слоя.  
На схеме архитектуры все выходные изображения делятся на два одинаковых участка — это связано с тем, что нейросеть обучалась на старых GPU GTX580, у которых было всего 3 ГБ видеопамяти. Для обработки использовались две видеокарты, чтобы параллельно выполнять операции над двумя частями изображения.

## ResNet
Основой сети ResNet является Residual-блок (остаточный блок) с shortcut-соединением, через которое данные проходят без изменений. Res-блок представляет собой несколько сверточных слоев с активациями, которые преобразуют входной сигнал x в F(x). Shortcut-соединение — это тождественное преобразование x -> x.

Создатели ResNet решили не складывать слои друг на друга для изучения отображения нужной функции напрямую, а использовать остаточные блоки, 
которые пытаются «подогнать» это отображение. Так ResNet стала первой остаточной нейронной сетью. 
Иначе говоря, она «перепрыгивает» через некоторые слои. Они больше не содержат признаков и используются для нахождения остаточной функции H(x) = F(x) + x вместо того, чтобы искать H(x) напрямую.

В результате экспериментов с ResNet выяснилось, что очень глубокие сети действительно можно обучить без ухудшения точности. Нейросеть достигла наименьшей ошибки в задачах классификации, которая превзошла даже человеческий результат.

## Non-maximum Suppression (NMS)
Алгоритм:  
1) Выбрать область с наибольшией уверенностью и запомнить его  
2) Найти IoU с оставшимися областями  
3) Удалить области с IoU больше заданного порога  
4) Повторять пп. 1-3, пока все области не отфильтруются

```python
import torch
from torchvision import models, transforms
from PIL import Image
import torch.nn.functional as F
import matplotlib.pyplot as plt
import cv2
import numpy as np


model = models.resnet50(pretrained = True)
model.eval()

W = model.fc.weight
b = model.fc.bias
print(W.shape)

activation = {}
def get_layer4(module, input, output):
    activation["layer4"] = output.detach()

handle = model.layer4.register_forward_hook(get_layer4)

transform = transforms.Compose([
    # transforms.Resize(256),
    # transforms.CenterCrop(224),
    # transforms.RandomRotation
    transforms.ToTensor(),
    transforms.Normalize(
       mean=[0.485, 0.456, 0.406],
        std=[0.229, 0.224, 0.225]
    )
])

img = Image.open("bird.jpeg").convert("RGB")
x = transform(img).unsqueeze(0)

with torch.no_grad():
    res = model(x)
    prob = F.softmax(res, dim=1)
    pred = prob.argmax(dim=1).item()

print(pred)

heat = activation["layer4"]
handle.remove()

w = W[pred].unsqueeze(0)
w = w.unsqueeze(2).unsqueeze(3)

print(heat.shape)
cam = F.relu((w * heat).sum(dim=1, keepdim = True))

cam_resized = F.interpolate(cam, size=(224, 224), mode="bilinear", align_corners=False)
cam_resized = cam_resized.squeeze().detach().cpu().numpy()  

cam_resized /= (cam_resized.max() + 1e-8)

heatmap = (cam_resized * 255).astype(np.uint8)
heatmap = cv2.applyColorMap(heatmap, cv2.COLORMAP_JET)[:, :, ::-1] 

img_orig = np.array(img.resize((224, 224)))


overlay = (0.7 * heatmap + 0.3 * img_orig).astype(np.uint8)

plt.figure(figsize=(10,4))
plt.subplot(1,3,1); plt.title("Image"); plt.imshow(img_orig); plt.axis("off")
plt.subplot(1,3,2); plt.title("CAM");   plt.imshow(cam_resized, cmap="jet"); plt.axis("off")
plt.subplot(1,3,3); plt.title("Overlay"); plt.imshow(overlay); plt.axis("off")
plt.tight_layout()
plt.show()
```

```python
import torch
import numpy as np
import cv2
from PIL import Image
from torchvision import models, transforms
import json
import urllib.request
import torch.nn.functional as F

model = models.resnet50(weights='DEFAULT')
model.eval()

features = []

def get_features_map(module, inputs, output):
    features.clear()
    features.append(output.data.cpu().numpy())

model.layer4.register_forward_hook(get_features_map)
weights = model.fc.weight.data.cpu().numpy()

url = "https://raw.githubusercontent.com/anishathalye/imagenet-simple-labels/master/imagenet-simple-labels.json"
response = urllib.request.urlopen(url)
imagenet_labels = json.loads(response.read())

preprocess = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
])

def process_frame(frame):
    img_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    img_pil = Image.fromarray(img_rgb)
    img_tensor = preprocess(img_pil).unsqueeze(0)
    
    with torch.no_grad():
        output = model(img_tensor)
    
    class_idx = np.argmax(output.data.numpy())
    class_name = imagenet_labels[class_idx]
    
    feature_maps = features[0][0] 
    w_k = weights[class_idx]
    
    cam = w_k.dot(feature_maps.reshape(2048, -1))
    cam = cam.reshape(feature_maps.shape[1], feature_maps.shape[2])
    
    cam = cam - np.min(cam)
    cam_max = np.max(cam)
    if cam_max != 0:
        cam = cam / cam_max
    cam = np.uint8(255 * cam)
    
    h, w, _ = frame.shape
    cam_large = cv2.resize(cam, (w, h))
    heatmap = cv2.applyColorMap(cam_large, cv2.COLORMAP_JET)
    
    result = cv2.addWeighted(frame, 0.6, heatmap, 0.4, 0)
    
    label = f"Class: {class_name}"
    cv2.putText(result, label, (20, 50), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 0), 4, cv2.LINE_AA)
    
    return result

def process_frame_multi_cam(frame):
    img_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    img_pil = Image.fromarray(img_rgb)
    img_tensor = preprocess(img_pil).unsqueeze(0)
    
    with torch.no_grad():
        output = model(img_tensor)
    
    probs = F.softmax(output, dim=1)[0].cpu().numpy()
    top3_indices = np.argsort(probs)[-3:][::-1] 
    top3_probs = probs[top3_indices]
    
    feature_maps = features[0][0] 
    h, w, _ = frame.shape
    
    result = np.zeros_like(frame)
    
    colors = [(0, 255, 0), (0, 0, 255), (255, 0, 0)]  
    
    THRESHOLD = 0.4
    
    for i, (class_idx, prob) in enumerate(zip(top3_indices, top3_probs)):
        class_name = imagenet_labels[class_idx]
        
        w_k = weights[class_idx]
        cam = w_k.dot(feature_maps.reshape(2048, -1))
        cam = cam.reshape(7, 7)
        
        cam = (cam - cam.min()) / (cam.max() - cam.min() + 1e-8)
        
        cam_binary = (cam > THRESHOLD).astype(np.uint8) * 255 
        
        mask_large = cv2.resize(cam_binary, (w, h), interpolation=cv2.INTER_NEAREST)
        
        color_intensity = mask_large.astype(np.float32) / 255.0 * 0.8  
        
        for c in range(3):
            result[:, :, c] = np.clip(
                result[:, :, c] + (color_intensity * colors[i][c]), 
                0, 255
            ).astype(np.uint8)
    
    y_offset = 50
    for i, (class_idx, prob) in enumerate(zip(top3_indices, top3_probs)):
        class_name = imagenet_labels[class_idx]
        label = f"{i+1}. {class_name}: {prob*100:.1f}%"
        cv2.putText(result, label, (20, y_offset), cv2.FONT_HERSHEY_SIMPLEX, 
                   0.7, colors[i], 2)
        y_offset += 35
    
    return result


def run_video_stream():
    cap = cv2.VideoCapture("dog.mp4")
    
    cap.set(cv2.CAP_PROP_FRAME_WIDTH, 1280)
    cap.set(cv2.CAP_PROP_FRAME_HEIGHT, 720)
    
    while True:
        ret, frame = cap.read()
        if not ret:
            break
            
        frame = cv2.resize(frame, (1280, 720))
        res_frame = process_frame(frame)
        
        cv2.imshow('video', res_frame)
        
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break
            
    cap.release()
    cv2.destroyAllWindows()

run_video_stream()
```
